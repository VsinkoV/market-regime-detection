## Visual Clusers of oil prices


Firstly,  we will downlaod the daily oil price history and turn it into a chart how volatile it has been over time. 
We are doing this because we are tryug to find the calmer periods over time and the more volatile periods and build things to detect them. 



```python
import yfinance
import numpy as np
import matplotlib.pyplot as plt
 
data = yfinance.download("WTI", start="2020-01-01", end="2026-05-25")
returns = np.log(data['Close'] / data['Close'].shift(1)).dropna()
plt.plot(returns)
```

Here we can see the days which have higher volity evidently with higher bars and other days with lower bars and further occur in these periods and how they occur in clusters.
